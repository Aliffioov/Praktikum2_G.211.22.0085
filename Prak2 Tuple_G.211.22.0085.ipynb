```python
# Membuat sebuah tuple
buah_tuple = ('apel', 'pisang', 'ceri', 'mangga')

# Mengakses elemen tuple
print(buah_tuple[0])  # Output: 'apel'

# Tuple bersifat tidak dapat diubah (immutable), jadi ini akan menghasilkan kesalahan
# buah_tuple[1] = 'jeruk'

# Menggunakan perulangan untuk mengakses elemen-elemen tuple
for buah in buah_tuple:
    print(buah)

# Mengetahui panjang tuple
panjang_tuple = len(buah_tuple)
print("Jumlah buah dalam tuple:", panjang_tuple)
```
