```python
# Membuat sebuah dictionary yang berisi data mahasiswa
mahasiswa = {
    'nama': 'Aliffio Ovata',
    'umur': 19,
    'jurusan': 'Teknik Informatika',
    'nim': 'G.211.22.0085'
}

# Mengakses nilai berdasarkan kunci
print("Nama:", mahasiswa['nama'])  # Output: 'Aliffio Ovata'
print("Umur:", mahasiswa['umur'])  # Output: 19

# Menambahkan pasangan kunci-nilai baru ke dalam dictionary
mahasiswa['alamat'] = 'JL Menjangan Barat No 28'

# Mengubah nilai yang ada dalam dictionary
mahasiswa['umur'] = 19

# Menghapus pasangan kunci-nilai dari dictionary
del mahasiswa['nim']

# Menggunakan perulangan untuk mengakses kunci dan nilai dalam dictionary
for key, value in mahasiswa.items():
    print(key, ":", value)

# Mengetahui panjang (jumlah pasangan kunci-nilai) dictionary
panjang_dict = len(mahasiswa)
print("Jumlah pasangan kunci-nilai dalam dictionary:", panjang_dict)
```
